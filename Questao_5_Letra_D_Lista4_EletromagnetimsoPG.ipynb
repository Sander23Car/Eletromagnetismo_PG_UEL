```python
#Importando as bibliotecas que ser√£o usadas

import numpy as np
import matplotlib.pyplot as plt
import math as mt
import scipy.special as scisp
```

# **Antena Linear**

```python
f     = np.pi
theta = np.arange(0,10,0.1)

ccostheta = np.cos(theta)
csentheta = np.sin(theta)
ang       = f*ccostheta
ccosang   = np.cos(ang)
ccosf     = np.cos(f)

r = ((ccosang - ccosf)**2)/(csentheta**4)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
plt.show()
```

# **Antena Circular**

```python
f     = np.pi

theta = np.arange(0,10,0.001)

termo = f*np.sin(theta)

r =(scisp.jv(1,termo))**2

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
plt.show()
```
