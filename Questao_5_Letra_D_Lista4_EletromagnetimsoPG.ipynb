```python
#Importando as bibliotecas que serão usadas

import numpy as np
import matplotlib.pyplot as plt
import math as mt
import scipy.special as scisp
```

```python
#Precisei colocar por que o código estava tentando dividir por zero
np.seterr(divide='ignore', invalid='ignore') 
```

# **Antena Linear**

```python
f     = np.pi
theta = np.arange(0,10,0.1)

ccostheta = np.cos(theta)
csentheta = np.sin(theta)
ang       = f*ccostheta
ccosang   = np.cos(ang)
ccosf     = np.cos(f)

r = ((ccosang - ccosf)**2)/(csentheta**4)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
plt.show()
```

# **Antena Circular**

```python
f     = np.pi

theta = np.arange(0,10,0.001)

termo = f*np.sin(theta)

r =(scisp.jv(1,termo))**2

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
plt.show()
```

# Qualitativamente, qual o alcance maior? Da antena circular ou linear?

Qualitativamente podemos dizer que o alcance da antena linear será melhor que o da antena circular, para cada caso analisado separadamente. Isso se deve ao fato de que temos muitos ruídos na antena circular. Entretanto podemos ter muita perda de sinal, pois as antenas lineares podem se propagar apenas na horizontal ou vertical, ao passo que antenas circulares se propagam nas duas direções ao mesmo tempo.
